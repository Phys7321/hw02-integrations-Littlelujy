```python
import numpy as np
from numpy import sqrt ,ones ,copy ,cos ,sin ,tan ,pi ,linspace ,array ,exp ,arange
from myint import * 
import matplotlib.pyplot as plt 
import scipy.integrate as sci

# x must not be greater than a 
# 100 samples to generate (how many number of points)
#The built-in functions are also coded to take either arrays or lambda functions (depending on the function) and arrays of where the y values are evaluated or limits.
x = np.linspace(0,4,100)
a = linspace(0.99,4,100)
Porig = np.vectorize(lambda x: 4./np.sqrt((2)*(a**4-x**4))) #don't need a for loop when using vectorize 
#write a as a vector and use "np.vectorize" to run different values for a
#we use (4./) to make sure the result is a float instead of an int.

#// Cubic Rule // 
Pcub = cubicrule(Porig,0,0.98,19)
p0, = plt.plot(x, Pcub, label= "Cubic Rule")
plt.legend([p0])
plt.show()

#// Trapezoid Rule// Most accurate??
Ptrap = trapezoidrule(Porig,0,0.98,19)
p1, = plt.plot(x,Ptrap,"C1", label = "Trapezoid Rule")
plt.legend([p1])
plt.show()

#//Errors in the Trapezoid Rule: 
# Approximation error of the trapezoid method is given by the Euler-Maclaurin formula
Ptraperr = trapezoiderr(Porig,0,0.98,19)
p2, = plt.plot(x,Ptraperr, "C2", label = "Error Trapezoid")
plt.legend([p2])
plt.show()

#// Simpson’s Rule // 
Psims = simpsonrule(Porig,0,0.98,19)
p3, = plt.plot(x,Psims, "C3", label = "Simpson's Rule")
plt.legend([p3])
plt.show()


#// Errors in Simpson’s Rule//
Psimserr = simpsonerr(Porig,0,0.98,19)
p4, = plt.plot(x,Psimserr, "C4", label = "Error Simpson's")
plt.legend([p4])
plt.show() 

#// Quartic Rule //
Pquar = quarticrule(Porig,0,0.98,19)
p5, = plt.plot(x,Pquar, "C5", label = "Quartic Rule")
plt.legend([p5])
plt.show() 

#// Romberg Rule// 
Promb = rombergrule(Porig,0,0.98,19)
p6, = plt.plot(x,Promb, "C6", label = "Romberg Rule")
plt.legend([p6])
plt.show() 

#



```
